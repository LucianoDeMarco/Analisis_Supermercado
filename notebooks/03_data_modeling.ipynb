```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Processed/Global_Superstore_Clean.csv")
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
customers = (
    df[
        ["Customer ID", "Customer Name", "Segment"]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

customers.head()
```

```python
customers["Customer ID"].nunique()
len(customers)
```

```python
products = (
    df[
        [
            "Product ID",
            "Product Name",
            "Category",
            "Sub-Category"
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

products.head()
```

```python
products["Product ID"].nunique()
len(products)
```

```python
orders = (
    df[
        [
            "Order ID",
            "Order Date",
            "Ship Date",
            "Ship Mode",
            "Order Priority"
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

orders.head()
```

```python
orders["Order ID"].nunique
len(orders)
```

```python
locations = (
    df[
        [
            "Country",
            "State",
            "City",
            "Region",
            "Market"
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)
```

```python
locations["Location ID"] = locations.index + 1
```

```python
order_details = df[
    [
        "Order ID",
        "Product ID",
        "Customer ID",
        "Sales",
        "Quantity",
        "Discount",
        "Profit",
        "Shipping Cost"
    ]
].copy()
```

```python
order_details = df.merge(
    locations,
    on=[
        "Country",
        "State",
        "City",
        "Region",
        "Market"
    ],
    how="left"
)
```

```python
order_details = order_details[
    [
        "Order ID",
        "Product ID",
        "Customer ID",
        "Location ID",
        "Sales",
        "Quantity",
        "Discount",
        "Profit",
        "Shipping Cost"
    ]
]
```

```python
customers.to_csv("../data/Processed/customers.csv", index=False)
products.to_csv("../data/Processed/products.csv", index=False)
orders.to_csv("../data/Processed/orders.csv", index=False)
locations.to_csv("../data/Processed/locations.csv", index=False)
order_details.to_csv("../data/Processed/order_details.csv", index=False)
```
